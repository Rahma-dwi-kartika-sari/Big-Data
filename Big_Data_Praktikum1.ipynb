```python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
df.show()

# Filtering data
df_filtered = df.filter(df['Usia'] > 30)
df_filtered.show()

# Menghitung rata-rata usia
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data berdasarkan usia
df_sorted = df.orderBy("Usia", ascending=False)
df_sorted.show()
```

**Tugas 5**

```python
import pandas as pd
from tabulate import tabulate

# Data awal di PySpark
data_spark = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data_spark, columns)

print("Data awal di PySpark")
df.show()

# Data tambahan di Pandas
data_pandas = {
    "Nama": ["Ali", "Budi", "Citra", "Dina"],
    "Hobi": ["Golf", "Golf", "Fotografi", "Tidur"],
    "Pekerjaan": ["CEO", "CTO", "Traveler", "Pelukis"]
}
df_pandas = pd.DataFrame(data_pandas)

print("\n Data tambahan di Pandas")
print(tabulate(df_pandas, headers='keys', tablefmt='psql', showindex=False))

# Mengonversi DataFrame dari PySpark ke Pandas
df_pandas_from_spark = df.toPandas()
print("\n Hasil konversi Spark -> Pandas")
print(tabulate(df_pandas_from_spark, headers='keys', tablefmt='psql', showindex=False))

# Gabungkan berdasarkan Nama (di Pandas)
df_pandas_merged = df_pandas_from_spark.merge(df_pandas, on="Nama")
print("\n Data gabungan di Pandas ")
print(tabulate(df_pandas_merged, headers='keys', tablefmt='psql', showindex=False))

# Mengonversi DataFrame dari Pandas ke PySpark
df_spark_from_pandas = spark.createDataFrame(df_pandas_merged)
print("\n Data gabungan di PySpark")
df_spark_from_pandas.show()

# Analisis sederhana: rata-rata usia
rata2 = df_spark_from_pandas.agg({"Usia": "avg"}).collect()[0][0]
print(f"\nRata-rata usia: {rata2:.2f}")
```

**Tugas 6**

```python
import pandas as pd
from tabulate import tabulate

# Data awal di PySpark
data_spark = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data_spark, columns)

print("Data awal di PySpark")
df.show()

# Data tambahan di Pandas
data_pandas = {
    "Nama": ["Ali", "Budi", "Citra", "Dina"],
    "Hobi": ["Golf", "Golf", "Fotografi", "Tidur"],
    "Pekerjaan": ["CEO", "CTO", "Traveler", "Pelukis"]
}
df_pandas = pd.DataFrame(data_pandas)

print("\nData tambahan di Pandas")
print(tabulate(df_pandas, headers='keys', tablefmt='psql', showindex=False))

# Mengonversi DataFrame dari PySpark ke Pandas
df_pandas_from_spark = df.toPandas()
print("\nHasil konversi Spark -> Pandas")
print(tabulate(df_pandas_from_spark, headers='keys', tablefmt='psql', showindex=False))

# Gabungkan berdasarkan Nama (di Pandas)
df_pandas_merged = df_pandas_from_spark.merge(df_pandas, on="Nama")
print("\nData gabungan di Pandas")
print(tabulate(df_pandas_merged, headers='keys', tablefmt='psql', showindex=False))

# Mengonversi DataFrame dari Pandas ke PySpark
df_spark_from_pandas = spark.createDataFrame(df_pandas_merged)
print("\nData gabungan di PySpark")
df_spark_from_pandas.show()

# Analisis sederhana: nilai maksimum usia
max_usia = df_spark_from_pandas.agg({"Usia": "max"}).collect()[0][0]
print(f"\nNilai maksimum usia: {max_usia}")
```
