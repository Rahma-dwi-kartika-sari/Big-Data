```python
import pandas as pd

df = pd.read_csv("/home/rahma/Documents/Titanic-Dataset.csv")
print("Dataset Awal:")
print(df.head())

print("\nMissing Value PerKolom:")
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler

num_cols = df.select_dtypes(include=['number']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
print("\nCek Missing Setelah Diisi Median:")
print(df[num_cols].isnull().sum())

scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
print("\nData Numerik Setelah Normalisasi:")
print(df[num_cols].head())
```

```python
from sklearn.preprocessing import LabelEncoder

cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna("Unknown")
print("Cek Missing Setelah Diisi pada Kolom Kategorikal:")
print(df[cat_cols].isnull().sum())

encoder = LabelEncoder()
for col in cat_cols:
    df[col] = encoder.fit_transform(df[col])
print("\nData Kategorikal Setelah Encoding:")
print(df[cat_cols].head())
```

```python
before = len(df)
df_cleaned = df.drop_duplicates()
after = len(df_cleaned)

print("Jumlah Data Sebelum:", before)
print("Jumlah Data Sesudah :", after)
print("Jumlah Duplikat Dihapus:", before - after)
```
